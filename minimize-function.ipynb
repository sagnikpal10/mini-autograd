### Minimizing a function using mini-autograd

```python
from autograd.tensor import Tensor
```

```python
# function
def sum_of_squares(x: Tensor) -> float:
    return (x * x).sum()
```

```python
x_1 = Tensor([10, -10, 10, -5, 6, 3, 1], requires_grad=True)
y = sum_of_squares(x_1)
print(y.data)
```

```python
x_2 = Tensor([11, -6, 12, -31, -1, 0, 1], requires_grad=True)
y = sum_of_squares(x_2)
print(y.data)
```

### For what value of x is the function sum_of_squares(x) will be minimum ?

```python
import numpy as np

x = Tensor([12, 0, 11, 5, 8, -9, 2], requires_grad=True) # initialization
print(sum_of_squares(x).data)
```

```python
y_vals = []

for i in range(20):
    x.zero_grad()

    y = sum_of_squares(x)
    y.backward()

    delta_x = Tensor(0.1) * x.grad
    x = x - delta_x

    y_vals.append(y.data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(y_vals) + 1), y_vals, marker='o', linestyle='-', color='b', label="sum_of_squares(x) per Iteration")
plt.xlabel("Iteration", fontsize=12)
plt.ylabel("sum_of_squares(x)", fontsize=12)
plt.title("Minimization of sum_of_squares (Convergence Plot)", fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
